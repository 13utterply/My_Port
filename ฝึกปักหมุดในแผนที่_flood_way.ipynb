```python
# ปรับปรุงข้อมูลคอลัมน์ gp_lat
# 1.ย้ายค่าจาก gp_lat → ไปไว้ใน gp_phone เฉพาะแถวที่ระบุ

# 2.แทนค่า gp_lat เดิมด้วย NaN เฉพาะแถวนั้น


import numpy as np
#copy df
df2 = df.copy()
# รายการ index ของแถวที่ต้องการย้าย
rows_to_move = [343, 344, 345, 289, 212, 233, 24, 86, 25]

# สร้างคอลัมน์ใหม่ 'gp_phone' ถ้ายังไม่มี
if 'gp_phone' not in df2.columns:
    df2['gp_phone'] = np.nan

# ย้ายข้อมูลจาก gp_lat ไปยัง gp_phone
df2.loc[rows_to_move, 'gp_phone'] = df2.loc[rows_to_move, 'gp_lat']

# แทนค่าที่ gp_lat ด้วย NaN
df2.loc[rows_to_move, 'gp_lat'] = np.nan

df2
```

```python
import folium
import pandas as pd

### folium.Marker() ไม่สามารถพล็อตจุดที่ไม่มีพิกัดได้ (จำเป็นต้องมีค่าตัวเลข)

# อ่านข้อมูลจากไฟล์ CSV
#df = pd.read_csv('locations.csv')
df2

# สร้างแผนที่ centered ที่ประเทศไทย (กรุงเทพฯ)
m = folium.Map(location=[13.75, 100.5], zoom_start=6)

# วนลูปเพิ่ม marker ตามพิกัด
for i, row in df2.iterrows():
    folium.Marker(
        location=[row['gp_lat'], row['gp_long']],
        popup=row['gp_name']
    ).add_to(m)

# บันทึกเป็น HTML เพื่อเปิดดูในเบราว์เซอร์
m.save('flood_way_thailand_map.html')
```

```python
# วิธีแก้ไข: กรองแถวที่ gp_lat และ gp_long ไม่เป็น NaN ก่อนวนลูป
df_valid = df2.dropna(subset=['gp_lat', 'gp_long'])

for i, row in df_valid.iterrows():
    folium.Marker(
        location=[row['gp_lat'], row['gp_long']],
        popup=row['gp_name']
    ).add_to(m)
    
# บันทึกเป็น HTML เพื่อเปิดดูในเบราว์เซอร์
# save ในไดรฟ์ C:\Users\PC
# m.save('flood_way_thailand_map.html') 

# save ในไดรฟ์ E:
m.save('E:/Datasets_OpenedSources/flood_way_thailand_map.html')
```

```python
for i, row in df_valid.iterrows():
    popup_text = f"""
        <b>{row['gp_name']}</b><br>
        ชนิด: {row['gp_type']}<br>
        เขต: {row['district']}<br>
        ระดับการควบคุม: {row['gp_water_control']}<br>
        ระดับการเตือน: {row['gp_warning']}<br>
        จุดวิกฤต: {row['gp_critical']}<br>
        โทรศัพท์: {row['gp_phone']}
    """

    folium.Marker(
        location=[row['gp_lat'], row['gp_long']],
        popup=folium.Popup(popup_text, max_width=300),
        tooltip=row['gp_name']  # ชื่อจะแสดงตอนเอาเมาส์ชี้
    ).add_to(m)
    
# save ในไดรฟ์ E:
m.save('E:/Datasets_OpenedSources/flood_way_thailand_map_popuptext.html')

webbrowser.open('E:/Datasets_OpenedSources/flood_way_thailand_map_popuptext.html')
```
